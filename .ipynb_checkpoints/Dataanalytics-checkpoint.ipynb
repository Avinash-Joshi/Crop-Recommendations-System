# Step 1 importing Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrices import r2_score
```

```python
df=pd.read_csv('Crop_recommendation.csv')
```

```python
## Cheking the number of unique values 
df.nunique()
```

```python
df["label"]=df["label"].replace({"rice":1,"maize":2,"chickpea":3,"kidneybeans":4,"pigeonpeas":5,"mothbeans":6,"mungbean":7,"blackgram":8,"lentil":9,"pomegranate":10,"banana":11,"mango":12,"grapes":13,"watermelon":14,"muskmelon":15,"apple":16,"orange":17,"papaya":18,"coconut":19,"cotton":20,"jute":21,"coffee":22})
```

```python
df.head()
```

```python
df.shape
```

```python
# Checking the number of null values in the dataframe
df.isnull().sum()
```

```python
# drop the rows according to column with low missing values
df.dropna(inplace=True)
df.isna().sum()
```

```python
X=df.drop('label',axis=1)
Y=df['label']
print(X.head(),end="\n\n")
print(Y.head())
print(df.head())
```

# Apllying Regreesion

```python
train_X,test_X,train_y,test_y=train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
reg=LinearRegression().fit(train_X,train_y)
```

```python
reg.score(train_X,train_y)
```

```python
reg.score(test_X,test_y)
```
